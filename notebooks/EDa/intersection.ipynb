```python
import pandas as pd
import geopandas as gpd
```

```python
grid = pd.read_csv("grid.csv")
intersection = pd.read_csv("../../data/intersection_data/intersection.csv")
```

```python
df_grid = grid
df_grid.drop(["geometry"], axis=1)
```

```python
df_grid["target"] = 0
```

```python
intersection = intersection.rename({"V5a": "target"}, axis=1).drop(["geometry", "V1"], axis=1).drop_duplicates()
```

```python
intersection.columns
```

```python
intersection = intersection.groupby('gid').apply(lambda group: group.sample(1)).reset_index(drop=True)
```

```python
df_grid = df_grid.drop(["geometry"], axis=1)
```

```python
missing_ids = pd.merge(intersection["gid"], df_grid, on='gid', how='right', indicator=True)
```

```python
missing_ids.head()
```

```python
result_df = missing_ids[missing_ids['_merge'] == 'right_only'].drop(['_merge', 'Unnamed: 0'], axis=1)
```

```python
multi_class_grid = pd.concat([intersection, result_df], ignore_index=True)
```

```python
len(multi_class_grid)
```

```python
multi_class_grid.to_csv("multi_class_grid.csv")
```
