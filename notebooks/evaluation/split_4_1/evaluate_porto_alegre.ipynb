# Evaluate XGBoost Multi Polos Result

Evaluation of each model of training polo

```python
import sys
sys.path.append('../../../src')
from utils import  generate_dataset_split, load_combinations
import pickle
from models.xgboost_model import XGBoostModel
from evaluator import XGBoostEvaluator
import pandas as pd
```

```python
addresses = []
for i in range(10):
    addresses.append(f'../../../data/models/split4_1/Porto Alegre/xgb_{i}.sav')
```

```python
combinations = load_combinations('models/split4_1/Porto Alegre/combination.pkl')
```

## Comb 0

```python
model_comb0 = XGBoostModel()
model_comb0.load_model(addresses[0])
```

```python
X0_val, y0_val = generate_dataset_split(combinations[0][1])
best_threshold = model_comb0.find_best_threshold(X0_val, y0_val)
```

```python
print(best_threshold)
```

```python
X_test, y_test  = generate_dataset_split(['Porto Alegre'])

eval_comb0 = XGBoostEvaluator(model_comb0)
eval_comb0.evaluate(X_test, y_test)
```

```python
eval_comb0.plot_roc_curve(y_test, "XGBOOST COMB 0")
```

```python
eval_comb0.plot_confusion_matrix(y_test)
```
