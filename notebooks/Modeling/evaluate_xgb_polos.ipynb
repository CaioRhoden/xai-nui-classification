# Evaluate XGBoost Multi Polos Result

Evaluation of each model of training polo

```python
import sys
sys.path.append('../../src')

import pickle
from utils import load_model
from evaluator import XGBoostEvaluator
import pandas as pd
```

## Porto Alegre

```python
polo1 = "Porto Alegre"
model_polo1 = load_model(f'../../data/models/polos/xgb_{polo1}.sav')
```

```python
X_test_polo1 = pd.read_pickle(f"../../data/model_input/X_test_{polo1}.pkl").drop(['Polo'], axis=1)
y_test_polo1= pd.read_pickle(f"../../data/model_input/y_test_{polo1}.pkl")

eval_polo1 = XGBoostEvaluator(model_polo1)
eval_polo1.evaluate(X_test_polo1, y_test_polo1)
```

```python
eval_polo1.plot_roc_curve(y_test_polo1, "XGBOOST PORTO ALEGRE")
```

```python
eval_polo1.plot_confusion_matrix(y_test_polo1)
```
