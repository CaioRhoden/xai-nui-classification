```python
import sys
sys.path.append('../../src')
import pandas as pd
from utils import  generate_dataset_split, load_combinations
import pickle
from evaluator import EnsambleEvaluator, XGBoostEvaluator
from models.xgboost_model import XGBoostModel
```

## Create Dataframe Basis

```python
basis = {
    'type': [],
    'porto_alegre': [],
    'brasilia': [],
    'maraba': [],
    'juazeiro_do_norte': [],
    'recife': [],
    'belo_horizonte': [],
    'auc_score': [],
    'f1_score': [],
    'acc': [],
    'recall': [],
    'precision': []
}

df = pd.DataFrame(basis)
```

```python
cities = ['Porto Alegre', 'Marabá', 'Brasília', 'Juazeiro do Norte', 'Recife', 'Belo Horizonte']
```

## Increment Splits 3-2

```python
for c in cities:
    combinations = load_combinations(f'models/split3_2/{c}/combination.pkl')
    print(c)
    for i in range(10):
        address = f'../../data/models/split3_2/{c}/xgb_{i}.sav'
        model = XGBoostModel()
        model.load_model(address)
        X_val, y_val = generate_dataset_split(combinations[i][1])
        best_threshold = model.find_best_threshold(X_val, y_val)
        X_test, y_test  = generate_dataset_split([c])
        eval = XGBoostEvaluator(model)
        eval.evaluate(X_test, y_test)
        new_line = eval.generate_csv_line('xgb_split3_2', combinations[i][0], combinations[i][1], [c])
        df = df.append(new_line, ignore_index=True)

        
```

## Increment Split 4_1

```python
df1 = df.copy()
```

```python
for c in cities:
    combinations = load_combinations(f'models/split4_1/{c}/combination.pkl')
    for i in range(5):
        address = f'../../data/models/split4_1/{c}/xgb_{i}.sav'
        model = XGBoostModel()
        model.load_model(address)
        X_val, y_val = generate_dataset_split(combinations[i][1])
        best_threshold = model.find_best_threshold(X_val, y_val)
        X_test, y_test  = generate_dataset_split([c])
        eval = XGBoostEvaluator(model)
        eval.evaluate(X_test, y_test)
        new_line = eval.generate_csv_line('xgb_split4_1', combinations[i][0], combinations[i][1], [c])
        df = df.append(new_line, ignore_index=True)
```

## Increment Ensembles

```python
df2 = df.copy()
```

```python
for c in cities:
    address = f'../../data/models/ensambles/{c}.sav'
    with open(address, 'rb') as file:
        model = pickle.load(file)
    X_test, y_test  = generate_dataset_split([c])
    eval = EnsambleEvaluator(model)
    eval.evaluate(X_test, y_test)
    new_line = eval.generate_csv_line('ensemble', [c])
    df = df.append(new_line, ignore_index=True)
```

## Save CSV
